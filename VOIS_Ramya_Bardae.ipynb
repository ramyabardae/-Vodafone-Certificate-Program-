# Task 4

## Find out the monthly revenue and analyze the findings

```python
# a.Create a column with Year and month field only give a suitable name to it
# Here we are using strftime() method that stands for string format time and can be applied to datetime objects.
df['year_month'] = df['Order Date'].dt.strftime('%Y-%m')
df
```

```python
# converting into pandas time object
df["year_month"] = pd.to_datetime(df["year_month"])
df.info()
```

```python
# b.Create a separate dataset from the data which will have two columns ,one which is created in step a of this task and second the monthly sales /revenue
df.set_index(df['year_month'],inplace=True)
monthly_sales = df['Sales'].resample('M').sum()
monthly_sales
```

```python
df.loc[:,'monthly_sales'] = df['Sales']
df.head(3)
```

```python
Monthly_sales = df[['year_month', 'monthly_sales']].copy()
Monthly_sales
```

```python
# plotting the data set
monthly_sales.plot(kind='line')
#Defining labels
plt.xlabel('year and month')
plt.ylabel('monthly revenue')
plt.title('Monthly Sales')

plt.show()
```

# Task 10

## Finding out the majority and minority customer cities on basis of


```python
# a. Number of customers
d2 = df[['City','Customer ID']].copy()
drop_dup = d2.drop_duplicates()
maxcity = d2['City']
print("Majority city based on no. of customers is: ",maxcity.max(),"\nMinority city based on no. of customers is: ",maxcity.min())


```

```python
# b. Sales value
d3 = df[['City','Sales']]
sumsale = d3.groupby(by='City').sum()
majcity = sumsale[sumsale.Sales == sumsale.Sales.max()]
minocity = sumsale[sumsale.Sales == sumsale.Sales.min()]
print("Majority city based on sales value: \n",majcity)
print("\n\nMinority city based on sales value: \n",minocity)

```

```python
# c. Number of quantity sold
d4 = df[['City','Quantity']]
sumquant = d4.groupby(by='City').sum()
qmajcity = sumquant[sumquant.Quantity == sumquant.Quantity.max()]
qminocity = sumquant[sumquant.Quantity == sumquant.Quantity.min()]
print("Majority city based on quantity: \n",qmajcity)
print("\n\nMinority city based on quantity: \n",qminocity)
```

# Task 11

## Find out the most and least sold product category from the store

```python
# a.Value based
d5 = df[['Category','Sales']]
catval = d5.groupby(by='Category').sum()
maxvalcat = catval[catval.Sales == catval.Sales.max()]
minvalcat = catval[catval.Sales == catval.Sales.min()]
print("Most sold category based on value: \n",maxvalcat)
print("\nLeast sold category based on value: \n",minvalcat)
```

```python
# b.Quantity based
d6 = df[['Category','Quantity']]
catsold = d6.groupby(by='Category').sum()
maxcatsold = catsold[catsold.Quantity == catsold.Quantity.max()]
mincatsold = catsold[catsold.Quantity == catsold.Quantity.min()]
print("Most sold category based on quantity: \n",maxcatsold)
print("\nLeast sold category based on quantity: \n",mincatsold)
```

# Task 12

## Find out the most and least sold product sub category from the store

```python
# Based on value
d7 = df[['Sub-Category','Sales']]
scatval = d7.groupby(by='Sub-Category').sum()
maxvalscat = scatval[scatval.Sales == scatval.Sales.max()]
minvalscat = scatval[scatval.Sales == scatval.Sales.min()]
print("Most sold sub-category based on value: \n",maxvalscat)
print("\nLeast sold sub-category based on value: \n",minvalscat)
```

```python
# Based on quantity
d8 = df[['Sub-Category','Quantity']]
scatsold = d8.groupby(by='Sub-Category').sum()
maxscatsold = scatsold[scatsold.Quantity == scatsold.Quantity.max()]
minscatsold = scatsold[scatsold.Quantity == scatsold.Quantity.min()]
print("Most sold sub-category based on quantity: \n",maxscatsold)
print("\nLeast sold sub-category based on quantity: \n",minscatsold)
```
